```python
df = pd.read_csv("생활폐기물(2020~2021).csv", encoding='cp949')
df.head()
```

```python
# 2020 ~ 2021년 폐합성수지류 발생현황 데이터
df1 = df[df['구분3']=='폐합성수지류']
df2 = df[df['구분4']=='폐합성수지류']

df_new = df1.append(df2)

df_new.drop(['구분2', '구분3', '구분4'], axis=1, inplace=True)
df_new.drop([24, 25, 26, 27], axis=0, inplace=True)
df_new = df_new.fillna(0)
df_new = df_new.groupby('구분1').sum()
df_new
```

```python
# 1996년 ~ 2019년 폐합성수지류 발생현황 데이터
df = pd.read_csv("생활폐기물(1996~2019).csv", encoding='cp949')
df1 = df[df['구분2'] == '폐합성수지류']
df1.drop(['구분2', '구분3', '구분4'], axis=1, inplace=True)
df1.drop(index=[5, 9, 32], axis=0, inplace=True)
df1 = df1.fillna(0)
df_old = df1.groupby('구분1').sum()
df_old
```

```python
# 전국 생활폐기물
life_all = pd.merge(df_old, df_new, how='inner', on='구분1')
life_all
life_all.to_csv("./life_all.csv", encoding="utf-8-sig")
```

---

```python
# 서울 생활폐기물
life_seoul = life_all[life_all.index == '서울']
life_seoul
life_seoul.to_csv("./life_seoul.csv", encoding="utf-8-sig")
```

```python
# 그 외 지역 생활폐기물
life_other = life_all[life_all.index != '서울']
life_other
life_other.to_csv("./life_other.csv", encoding="utf-8-sig")
```
