```python
import pandas as pd

# 데이터 불러오기
data = pd.read_csv("/content/drive/MyDrive/Ecothon2023/data/collect.csv")

# 전체 데이터의 type이 "수거공간"이므로 "type" 열 제거
data = data.drop(["type"], axis=1)

# tell -> contact 컬럼명 변경
data = data.rename(columns={"tell": "contact"})

# 결과 확인
data
```

```python
# html2text 라이브러리 설치 
!pip install html2text
```

```python
from html2text import html2text

# HTML 태그를 제거하는 함수
def convert_html_to_text(html):
  text = html2text(html)
  # "\n" 제거
  text = text.replace("\n", "")
  # "_" 제거
  if "_" in text:
    text = text.replace("_", "")
  # contact 컬럼 전화번호만 추출
  if "tel" in text:
    start_index = text.find(":") + 1
    end_index = text.find(")")
    text = text[start_index:end_index]
  return text

# data의 모든 컬럼에 대해 함수 적용
data = data.applymap(convert_html_to_text)

# 결과 확인
data
```

```python
# 이름이 "mydata.csv"인 csv 파일로 인덱스 제외하고 저장
data.to_csv("mydata.csv", index=False)
```
