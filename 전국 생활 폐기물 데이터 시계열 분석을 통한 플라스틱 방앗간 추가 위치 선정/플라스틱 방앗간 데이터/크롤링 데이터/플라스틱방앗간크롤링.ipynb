# Scraping

### 수거공간

```python
collect_pd=pd.DataFrame()
collect_pd
```

```python
excutable_path = "./chromedriver.exe"
for i in range(1,8):
    source_url = "https://ppseoul.com/map/?sort=TIME&category=W5B0v5yK42&keyword_type=all&page="+str(i)
    driver = webdriver.Chrome(executable_path=excutable_path)
    driver.get(source_url)
    
    # 수거공간 정보 긁어오기
    req = driver.page_source
    soup = BeautifulSoup(req, "html.parser")
    collect_info = soup.find_all("div", {"class":"map_contents inline-blocked"})

    # 수거공간 정보 데이터프레임에 종류별로 저장
    for j in range(10):
        collect=collect_info[j]
        collect_title=collect.find("em",{"id":"comment_writer"})
        collect_address=collect.find("p",{"class":"adress"})
        collect_tell=collect.find("p",{"class":"tell"})
        collect_date=collect.find("em",{"class":"comment_wtime"})
        collect_pd=collect_pd.append({'title':collect_title,'address':collect_address,
                                  'tell':collect_tell,'date':collect_date,'type':'수거공간'},
                                     ignore_index=True)
```

### 작업공간

```python
workspace_pd=pd.DataFrame()
workspace_pd
```

```python
excutable_path = "./chromedriver.exe"
for i in range(1,4):
    source_url = "https://ppseoul.com/map/?sort=TIME&category=CXU7D52EUf&keyword_type=all&page="+str(i)
    driver = webdriver.Chrome(executable_path=excutable_path)
    driver.get(source_url)
    
    # 작업공간 정보 긁어오기
    req = driver.page_source
    soup = BeautifulSoup(req, "html.parser")
    workspace_info = soup.find_all("div", {"class":"map_contents inline-blocked"})

    # 작업공간 정보 데이터프레임에 종류별로 저장
    # 3페이지만 정보가 2개이므로 코드 분리
    if i==3: 
        for j in range(2):
            workspace=workspace_info[j]
            workspace_title=workspace.find("em",{"id":"comment_writer"})
            workspace_address=workspace.find("p",{"class":"adress"})
            workspace_tell=workspace.find("p",{"class":"tell"})
            workspace_date=workspace.find("em",{"class":"comment_wtime"})
            workspace_pd=workspace_pd.append({'title':workspace_title,'address':workspace_address,
                                      'tell':workspace_tell,'date':workspace_date,'type':'작업공간'},
                                         ignore_index=True)
    else:
        for j in range(10):
            workspace=workspace_info[j]
            workspace_title=workspace.find("em",{"id":"comment_writer"})
            workspace_address=workspace.find("p",{"class":"adress"})
            workspace_tell=workspace.find("p",{"class":"tell"})
            workspace_date=workspace.find("em",{"class":"comment_wtime"})
            workspace_pd=workspace_pd.append({'title':workspace_title,'address':workspace_address,
                                      'tell':workspace_tell,'date':workspace_date,'type':'작업공간'},
                                         ignore_index=True)
```
